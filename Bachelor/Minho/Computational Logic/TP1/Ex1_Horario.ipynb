# Problema: 

1. Pretende-se construir um hor√°rio semanal para o plano de reuni√µes de projeto de uma ‚ÄúStartUp‚Äù de acordo com as seguintes condi√ß√µes:
    1. Cada reuni√£o ocupa uma sala (enumeradas **1**..**S**) durante um ‚Äúslot‚Äù  **1**..**T** (hora,dia). 
    2.  Cada reuni√£o tem associado um projeto (enumerados **1..P**) e um conjunto de participantes. Os diferentes colaboradores s√£o enumerados **1**..**C**.
    3. Cada projeto tem associado um conjunto de colaboradores, dos quais um  √© o l√≠der. Cada projeto realiza um dado n√∫mero de reuni√µes semanais. 
    4. O l√≠der do projeto participa em todas as reuni√µes do seu projeto; os restantes colaboradores podem ou n√£o participar consoante a sua disponibilidade, num m√≠nimo (‚Äúquorum‚Äù) de  50\% do total de colaboradores do projeto.<br>

    


S√£o ‚Äúinputs‚Äù do problema:<br>
    i. Os par√¢metros **S,T,P,C** <br>
    ii. O conjunto de colaboradores de cada projeto, o seu l√≠der e o n√∫mero de reuni√µes semanais.<br>
    iii. A disponibilidade de cada participante, incluindo o lider. Essa disponibilidade   √© um conjunto de ‚Äúslots‚Äù  representada numa matriz booleana de acessibilidade com uma linha por cada participante **1**..**C** e uma coluna por ‚Äúslot‚Äù **1**..**T**.<br><br>
S√£o crit√©rios de optimiza√ß√£o:<br>
    i. Maximizar o n√∫mero de reuni√µes efetivamente realizadas<br>
    ii. Minimizar o n√∫mero m√©dio de reuni√µes por participante.

# An√°lise do Problema

Este √© um problema de aloca√ß√£o que requer uma matriz boleana para se resolver. Como tal, √© necess√°rio a utiliza√ß√£o de variaveis bin√°rias (1 ou 0).<br><br>
Para resolver este problema ser√° necess√°rias S √ó T √ó D √ó P √ó C vari√°veis.<br>
Estas vari√°veis pertecer√£o √† seguinte familia: $x_{s,t,d,p,c}$.<br><br>
$x_{s,t,d,p,c}=1 \iff$ Colaborador estar√° presente na reuni√£o p, sala s e no slot (t,d)<br><br>
Algumas vari√°veis que useramos para descrever as limita√ß√µes e as obriga√ß√µes do nosso problema:
* $s ‚àà ùêí = \{1..S\}; \rightarrow$ (Salas)
* $t ‚àà ùêì = \{1..T\}; \rightarrow$ (Slots de tempo = Horas)
* $d ‚àà ùêÉ = \{1..D\}; \rightarrow$ (Dias)
* $p ‚àà ùêè = \{1..P\}; \rightarrow$ (Projetos)
* $c ‚àà ùêÇ = \{1..C\}; \rightarrow$ (Colaboradores)
* $y_p \in ùêÇ \rightarrow$ (Lider do projeto)
* $n_p \rightarrow$ (Numero de reuni√µes semanais)
* $X_p \subseteq ùêÇ \rightarrow$ (Grupo dos colaboradores do projeto)
* $S_p \rightarrow$ (Lista de Slots de tempo disponiveis para um colaborador espec√≠fico)



# Limita√ß√µes e obriga√ß√µes

**1. As salas s√≥ podem ter uma reuni√£o a decorrer de cada vez**<br>
√â necess√°rio que o somat√≥rio de todos os projetos para uma espec√≠fica sala e num espec√≠fico hor√°rio, seja no m√°ximo 1, assim garantimos que cada sala s√≥ pode ter 1 uma reuni√£o a decorrer.
$$\forall_{s \in ùêí}\forall_{t \in ùêì}\forall_{d \in ùêÉ}:\sum_{p=1}^{ùêè} x_{s,t,d,p,y_p} \leq 1 $$
**2. O lider tem de ter participado em todas as reuni√µes**<br>
√â necess√°rio que o somat√≥rio de todos os hor√°rios e salas de um projeto p seja igual √† sua pre-definida quantidade de reuni√µes $(n_p)$.
$$‚àÄ_{p \in ùêè} : \sum_{s=1}^{ùêí}\sum_{t=1}^{ùêì}\sum_{d=1}^{ùêÉ} x_{s,t,d,p,y_p} = n_p$$
**3. Colaboradores s√≥ podem participar se tiverem disponibilidade**<br>
√â necess√°rio que numa sala, hor√°rio e projeto espec√≠fico, se o colaborador em quest√£o n√£o tenha esse hor√°rio dispon√≠vel, ent√£o n√£o poder√° participar.
$$\forall_{s \in ùêí}\forall_{t \in ùêì}\forall_{d \in ùêÉ}\forall_{p \in ùêè}\forall_{c \in ùêÇ}:(t,d) \notin S_p \implies x_{s,t,d,p,c} = 0 $$
**4. Os Colaboradores s√≥ podem participar numa reuniao de cada vez**<br>
√â necess√°rio que num hor√°rio, colaborador e projeto  espec√≠fico, o som√°torio de todas as salas seja no m√°ximo 1, assim garantimos que esse espec√≠fico colaborador n√£o estar√° noutra sala.
$$\forall_{t \in ùêì}\forall_{d \in ùêÉ}\forall_{p \in ùêè}\forall_{c \in ùêÇ}:\sum_{s=1}^{ùêí} x_{s,t,d,p,c} \leq 1$$
**5. Colaboradores n√£o podem participar em projetos que n√£o sejam dele**<br>
√â necess√°rio que numa sala, hor√°rio e projeto espec√≠fico, se o colaborador em quest√£o n√£o estiver na lista dos colaboradores desse projeto espec√≠fico, ent√£o nao pode participar.
$$\forall_{s \in ùêí}\forall_{t \in ùêì}\forall_{d \in ùêÉ}\forall_{p \in ùêè}\forall_{c \in ùêÇ}:c \notin X_p \implies x_{s,t,d,p,c} = 0 $$
**6. No minimo, √© necess√°rio 50% dos colaboradores nas reunioes ("quorum")**<br>
√â necess√°rio que num hor√°rio e projeto  espec√≠fico, o som√°torio de todas os colaboradores desse projeto tem de ser superior a 50% dos colaboradores alocados.<br>
**NOTA:** Foi usado a express√£o ($x_{s,t,d,p,y_p} √ó \frac{|X_p|}{2}$) porque assim, com a vari√°vel ($x_{s,t,d,p,y_p}$) conseguimos controlar a existencia ou n√£o dessa reuni√£o (0 se n√£o existir, 1 se existir). Com a aus√™ncia dessa vari√°vel, estariamos a partir do pressuposto que a reuni√£o existe sempre.
$$\forall_{s \in ùêí}\forall_{t \in ùêì}\forall_{d \in ùêÉ}\forall_{p \in ùêè}:\sum_{c=1}^{X_p}x_{s,t,d,p,c} \geq x_{s,t,d,p,y_p} √ó \frac{|X_p|}{2} $$ 
**7. O lider √© obrigat√≥rio na reuni√£o**<br>
√â necess√°rio que num hor√°rio e projeto  espec√≠fico, o som√°torio de todas os colaboradores desse projeto tem de ser no m√°ximo igual ao somat√≥rio dos colaboradores.<br>
**NOTA:** Foi usado a express√£o ($x_{s,t,d,p,y_p} √ó |X_p|)$ porque assim, com a vari√°vel ($x_{s,t,d,p,y_p}$) conseguimos controlar a exist√™ncia do l√≠der nela (0 se n√£o existir, 1 se existir).
$$\forall_{s \in ùêí}\forall_{t \in ùêì}\forall_{d \in ùêÉ}\forall_{p \in ùêè}:\sum_{c=1}^{X_p}x_{s,t,d,p,c} \leq x_{s,t,d,p,y_p} √ó |X_p| $$

**Resolver o codigo**
1. √â definido a fun√ß√£o "horario" com return de: 0 ou um dicion√°rio com todas as reuni√µes
    1. Importar o Solver
    2. Criar uma Lista de tuplos com o tipo do solver.BoolVar
    3. Adicionar ao solver todas Limita√ß√¥es e Obriga√ß√µes
    4. Executar o solver
    

```python
def horario(S, T, P, C, dados, disponibilidadeColabs):

    # Importar o solver
    solver = pywraplp.Solver.CreateSolver("SCIP")

    # Criar uma Lista de tuplos com o tipo do solver.BoolVar
    listVariaveis = {}
    D = 5
    for s in range(1, S+1):
        for t in range(1, T+1):
            for d in range(1, D+1):
                for p in range(1, P+1):
                    for c in range(1, C+1):
                        listVariaveis[s, t, d, p, c] = solver.BoolVar(
                            "s[%i]t[%i]d[%i]p[%i]c[%i]" % (s, t, d, p, c))

    # Limita√ß√£o/Obriga√ß√£o 1
    # As salas s√≥ podem ter uma reuni√£o a decorrer de cada vez
    for s in range(1, S+1):
        for t in range(1, T+1):
            for d in range(1, D+1):
                solver.Add(
                    sum(listVariaveis[s, t, d, p, dados[p]["lider"]] for p in range(1, P+1)) <= 1)

    # Limita√ß√£o/Obriga√ß√£o 2
    # O l√≠der √© obrigado a estar presente nas reunioes todas
    for p in range(1, P+1):
        solver.Add(sum(listVariaveis[s, t, d, p, dados[p]["lider"]] for t in range(
            1, T+1) for d in range(1, D+1) for s in range(1, S+1)) == dados[p]["nReunioesSemanais"])

    # Limita√ß√£o/Obriga√ß√£o 3
    # Colaboradores s√≥ podem participar se tiverem disponibilidade
    for s in range(1, S+1):
        for t in range(1, T+1):
            for d in range(1, D+1):
                for p in range(1, P+1):
                    for c in range(1, C+1):
                        if (t, d) not in disponibilidadeColabs[c]:
                            solver.Add(listVariaveis[s, t, d, p, c] == 0)

    # Limita√ß√£o/Obriga√ß√£o 4
    # Os Colaboradores s√≥ podem participar numa reuniao de cada vez
    for t in range(1, T+1):
        for d in range(1, D+1):
            for p in range(1, P+1):
                for c in range(1, C+1):
                    solver.Add(sum(listVariaveis[s, t, d, p, c]
                               for s in range(1, S+1)) <= 1)

    # Limita√ß√£o/Obriga√ß√£o 5
    # Colaboradores n√£o podem participar em projetos que n√£o sejam dele
    for s in range(1, S+1):
        for t in range(1, T+1):
            for d in range(1, D+1):
                for p in range(1, P+1):
                    for c in range(1, C+1):
                        if c not in dados[p]["colaboradores"]:
                            solver.Add(listVariaveis[s, t, d, p, c] == 0)

    # Limita√ß√£o/Obriga√ß√£o 6
    # No minimo, √© necess√°rio 50% dos colaboradores nas reunioes ("quorum")
    for s in range(1, S+1):
        for t in range(1, T+1):
            for d in range(1, D+1):
                for p in range(1, P+1):
                    solver.Add(sum(listVariaveis[s, t, d, p, c] for c in dados[p]["colaboradores"]) >=
                               (listVariaveis[s, t, d, p, dados[p]["lider"]]*(len(dados[p]["colaboradores"])/2)))

    # Limita√ß√£o/Obriga√ß√£o 7
    # O lider √© obrigatorio na reuniao
    for s in range(1, S+1):
        for t in range(1, T+1):
            for d in range(1, D+1):
                for p in range(1, P+1):
                    solver.Add((sum(listVariaveis[s, t, d, p, c] for c in dados[p]["colaboradores"]))
                               <= listVariaveis[s, t, d, p, dados[p]["lider"]] * len(dados[p]["colaboradores"]))

    # Executar o solver
    status = solver.Solve()
    if status == solver.OPTIMAL:
        final = []
        for s in range(1, S+1):
            for t in range(1, T+1):
                for d in range(1, D+1):
                    for p in range(1, P+1):
                        auxListColaboradores = []

                        # Retornar os dados finais para usar posteriormente numa tabela
                        for c in range(1, C+1):
                            if listVariaveis[s, t, d, p, c].solution_value():
                                auxListColaboradores.append(c)
                        if auxListColaboradores:
                            #final = [((projeto,dia),Dicionario com dados).....]
                            #O projeto e o dia foram colocados no inicio para um sort na fun√ß√£o seguinte
                            final.append(((p,d),{
                                "projeto": p,
                                "horario": (t, d),
                                "lider": dados[p]["lider"],
                                "sala": s,
                                "colaboradores": sorted(auxListColaboradores)
                            }))
        return final
    else:
        return 0
```

**Fun√ß√µes para gerar parametros**

```python
# 2 Dicionarios e 1 Set pra guardar toda a informa√ß√£o essencial
disponibilidadeColabs = {}
dadosProjetos = {}
slots = set()

# Esta fun√ß√£o permite guardar todos os dados de cada projeto
def adicionar_dados_projetos(nr, colabs, lider, nrReunioesSemanais):
    dadosProjetos[nr] = {
        "colaboradores": colabs,
        "lider": lider,
        "nReunioesSemanais": nrReunioesSemanais}

# Esta funcao cria todas as slots disponiveis
def todas_slots(T):
    for t in range(1, T+1):
        for d in range(1, 6):
            if (t, d) not in slots:
                slots.add((t, d))
```
